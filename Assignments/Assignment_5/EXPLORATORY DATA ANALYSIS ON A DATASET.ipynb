# 1. Data Cleaning and Preparation

```python
#Load the Dataset
```

```python
import pandas as pd

# Corrected raw URL from GitHub
url = "https://raw.githubusercontent.com/snehuuu28/Excelr_Data_Science/main/Assignments/Assignment_5/Cardiotocographic.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(url)
```

```python
#Handling Missing Value
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
# Impute or handle missing values
df.fillna(df.median(), inplace=True)  # Example imputation
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values
```

```python
#Check and Correct Datatypes
```

```python
# Convert columns to appropriate data types
df['LB'] = pd.to_numeric(df['LB'], errors='coerce')
```

```python
df
```

```python
#Detect and Treat Outliers
```

```python
import matplotlib.pyplot as plt

# List all columns in the DataFrame
all_columns = df.columns

# Number of columns for plotting
num_cols = 2
num_rows = (len(all_columns) + num_cols - 1) // num_cols  # Calculate rows needed

# Create subplots
plt.figure(figsize=(10, 3 * num_rows))

for i, col in enumerate(all_columns):
    plt.subplot(num_rows, num_cols, i + 1)  # Grid of num_rows x num_cols
    plt.boxplot(df[col].dropna())
    plt.title(f'Boxplot for {col}')
    plt.xlabel(col)
    plt.ylabel('Values')

# Adjust layout
plt.tight_layout()
plt.show()
```

# 3.Data Visualization

```python
#Distribution of Numerical Variables
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')

# Determine the number of numerical columns
num_columns = len(df.select_dtypes(include=['number']).columns)

# Calculate the number of rows and columns for the subplot grid
ncols = 4  # Number of columns in the grid
nrows = (num_columns + ncols - 1) // ncols  # Calculate number of rows needed

# Plot histograms for all numerical columns
plt.figure(figsize=(10, 4 * nrows))
for i, column in enumerate(df.select_dtypes(include=['number']).columns):
    plt.subplot(nrows, ncols, i + 1)
    sns.histplot(df[column].dropna(), kde=False)
    plt.title(column)
plt.tight_layout()
plt.show()
```

```python
#Relationships Between Variables
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pair plot for all numerical columns
sns.pairplot(df.select_dtypes(include=['number']))
plt.suptitle('Pair Plot of All Numerical Columns', y=1.02)
plt.show()
```

```python
# Scatter plot between 'LB' and 'AC'
plt.figure(figsize=(8, 6))
sns.scatterplot(x='LB', y='AC', data=df)
plt.title('Scatter Plot of LB vs. AC')
plt.xlabel('LB')
plt.ylabel('AC')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix
corr = df.corr()

# Plot correlation heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of All Numerical Columns')
plt.tight_layout()
plt.show()
```

```python
print(corr)
```

```python
import numpy as np
# Create a boolean mask for the upper triangle of the correlation matrix
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set diagonal to False (self-correlation)
mask[np.diag_indices_from(mask)] = False

# Find the highly correlated pairs
high_corr_pairs = corr[mask].stack().sort_values(ascending=False)
threshold = 0.7  # Define your threshold for high correlation

# Filter pairs with correlation above the threshold
high_corr_pairs = high_corr_pairs[high_corr_pairs > threshold]
print("Highly Correlated Pairs:")
print(high_corr_pairs)
```
