# BASIC STATISTICS

In this analysis, we'll perform descriptive analytics on a Sales & Discounts dataset. The goal is to explore the data through statistical measures and visualizations, and to preprocess the dataset for subsequent machine learning models. We’ll start by examining the numerical and categorical variables, then proceed with data visualization, standardization, and converting categorical variables into dummy variables.

```python
#Descriptive Analytics for Numerical Columns
```

* Objective: Compute and analyze basic statistical measures for numerical columns.
* Steps:
      Load the Dataset:
      Import the dataset using Python's Pandas library.

```python
import pandas as pd

# URL of the raw CSV file
url = "https://raw.githubusercontent.com/snehuuu28/Excelr/main/Assignments/Assignment_1/sales_data_with_discounts.csv"

# Load the dataset into a DataFrame
df = pd.read_csv(url)
```

```python
df
```

```python
# Display the first few rows of the dataset
df.head()
```

```python
#Identify Numerical Columns:

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
```

```python
numerical_cols
```

```python
#Calculate Basic Statistical Measures:
#Compute the mean, median, mode, and standard deviation for each numerical column.

df.describe()
```

* Interpretation:

1) Mean provides the average value.

2) Median shows the middle value when data is sorted.

3) Mode indicates the most frequent value.

4) Standard Deviation measures data spread. High values indicate more variability, while low values suggest data points are closer to the mean.

# Data Visualization

Objective: Visualize the distribution and relationships of numerical and categorical variables.

# Histogram

* Steps:
  
1) Plot histograms for each numerical column to visualize the distribution.
2) Analyze skewness and the presence of outliers.

```python
import matplotlib.pyplot as plt

df.hist(bins=20, figsize=(15, 10))
plt.show()
```

* Interpretation:

1) Skewness indicates whether data is symmetrically distributed.

2) Outliers are extreme values that differ significantly from the majority of data.

# Boxplot

* Steps: 
1) Create boxplots to identify outliers and the interquartile range (IQR).
2) Analyze findings, focusing on extreme values and unusual distributions

```python
df.plot(kind='box', subplots=True, layout=(3,3), figsize=(15, 10))
plt.show()
```

* Interpretation:
1) IQR represents the middle 50% of the data.

2) Outliers are shown as points outside the whiskers of the boxplot.

# Bar Chart Analysis for Categorical Columns:

* Steps:
1) Identify categorical columns using df.select_dtypes(include=[‘object’]).
2) Create bar charts to visualize the frequency or count of each category.

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    df[col].value_counts().plot(kind='bar', title=col)
    plt.show()
```

* Interpretation:
1) Understand the distribution of categories and identify dominant categories.

# Standardization of Numerical Variables

* Objective: Scale numerical variables for uniformity, improving dataset suitability for analytical models.

* Steps:
1) Concept of Standardization:
    Standardization involves transforming data to have a mean of 0 and a standard deviation of 1 using the z-score formula: 
         𝑧 = 𝑥−𝜇/𝜎

2) Standardize Numerical Columns:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_standardized = pd.DataFrame(scaler.fit_transform(df[numerical_cols]), columns=numerical_cols)
```

```python
# Compare Before and After:
# Plot histograms before and after standardization to observe the effect.
```

```python
import matplotlib.pyplot as plt

data_standardized.hist(bins=15, figsize=(15, 10))
plt.show()
```

# Conversion of Categorical Data into Dummy Variables

* Objective: Transform categorical variables into a format suitable for machine learning algorithms.
* Steps:
1) Need for One-Hot Encoding: ML models require numerical inputs. One-hot encoding converts categorical data into binary columns.
2) Apply One-Hot Encoding:
3) Display Transformed Dataset:

```python
#Apply One-hot Encoding
df_encoded = pd.get_dummies(df, columns=categorical_cols)
```
